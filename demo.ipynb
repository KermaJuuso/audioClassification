This is a demo used for testing the classifying pipeline.
In this file one audio file can be inputted and it goes through the whole pipeline:
demo_data -> preprocess.py -> processed_data -> extract_features.py -> feature.csv -> classifying.py/ready_made_model

1. Start with a demo_data folder with one audio file (or as many as you like)

```python
import pandas as pd
import joblib
from components.preprocess import audio_data
from components.extract_features import extract_all
import numpy as np
```

```python
destination_path = "processed_demo_data"
def demo(demo_folder_path):
    processed_data = audio_data(demo_folder_path, "demo", destination_path)

    row = extract_all(destination_path, label="demo")
    
    df = pd.DataFrame(data=row)
    df.to_csv('demo_features.csv', mode='w', index=False)

    features = df.iloc[:, 2:].to_numpy().astype(np.float32)

    model = joblib.load("components/random_forest_model.joblib")
    pred = model.predict(features)
    mapping = {0: "car", 1: "tram"}
    print("Prediction:", mapping[pred[0]])
```

```python
demo("demo_data")  # Simply run this line when you have your path to a folder with a single audio file
```
