```python
import numpy as np
import matplotlib.pyplot as plt
import librosa as lb
from scipy.fftpack import fft, ifft
import sounddevice as sd
```

```python
audio, sr = lb.load("1_bus.wav", sr=None)
```

```python
t = np.arange(len(audio)) / sr
plt.plot(t, audio)# Can downsample with [::number]
plt.title("Audio")
plt.xlabel("Seconds")
plt.ylabel("Amplitude")
plt.grid(True)
plt.show()
```

```python
hop_len = int(sr * (0.05))
win_len = int(sr * 0.1)

X = lb.stft(audio, n_fft=win_len, hop_length=hop_len, win_length=win_len, window='hamming')
```

```python
Xabs = np.abs(X)
db = lb.amplitude_to_db(Xabs, ref=np.max)

plt.figure(figsize=(10,5))
lb.display.specshow(db, sr=sr, y_axis='log', x_axis='time')
plt.colorbar(label='Power (dB)')
plt.title('Librosa spectrogram')
plt.show()
```
